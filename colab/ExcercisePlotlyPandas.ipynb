```python
# matplotlib 
# plotly 
# plotly-express
# !pip install plotly-express

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv")

import plotly.express as px
fig = px.line( df ,x="data", y=["totale_positivi","dimessi_guariti"], title="Klkl")
fig.update_layout(
    title="Plotly express",
    paper_bgcolor="blue",
    plot_bgcolor="black",
    hovermode ="x"
)
fig.show()

import plotly.graph_objects as go
fig = go.Figure()
fig.add_traces ( go.Scatter( x=df["data"], y=df["totale_positivi"], name="totale_positivi"))
fig.add_traces ( go.Scatter( x=df["data"], y=df["dimessi_guariti"], name="totale_positivi"))
fig.update_layout(
    title="Plotly",
    paper_bgcolor="red",
    plot_bgcolor="black",
    hovermode ="x"
)
fig.show()

fig.write_html("file.html")
fig.write_image("fig.svg")
fig.write_image("fig.pdf")
fig.write_image("fig.png")
#fig.write_image("fig.eps")

#import matplotlib.pyplot as plt
#plt.plot(df["data"],df["totale_positivi"])
#plt.plot(df["data"],df["dimessi_guariti"])
#plot.show()
```

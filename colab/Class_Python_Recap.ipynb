```
import pandas as pd

def get_data(i, c, p, o=4, test=2):
  t = pd.read_csv("https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv")
  l = ["deceduti","totale_positivi","dimessi_guariti","totale_casi"]
  index = 110
  print(f"data: {t['data'].values[index]}")
  for el in l:
    print("-------------------")
    print(f"{el}: {t[el].values[index]}")
    print(f"{el} incremento: {t[el].values[index-1]-t[el].values[index]}")
  return 4, [4,5,5], (4,4)

a, b, c = get_data(5,6,4)
print(a,b,c)
#display(t)
#print(t.keys())
#print(t.keys())

#print("totale positivi oggi: ", list(t["totale_positivi"])[-1])
#np.abs(t["totale_positivi"].values[-1]-t["totale_positivi"].values[-2])
```

```
import schedule
import time

schedule.every(5).seconds.do(get_data)
#schedule.every(10).minutes.do(job)
#schedule.every().hour.do(job)
#schedule.every().day.at("19:52").do(job)

while 1:
    schedule.run_pending()
    time.sleep(1)
```

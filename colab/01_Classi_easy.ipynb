```python
class Persona:
    def __init__(self,a,b):
        self.nome = a
        self.eta = b

daniele = Persona('Daniele',36)
daniele.nome
```

```python
class Persona:
    def __init__(self,a,b):
        self.nome = a
        self.eta = b

    def double(self,x):
        ''' moltiplica l'input x2 volte'''
        return x*2

    def triple(self,x):
        ''' moltiplica l'input x3 volte'''
        return x*3

    def quadratoanni(self):
        return self.eta**2

daniele = Persona('Daniele',36)
daniele.nome
```

```python
daniele.double(23)
```

```python
marco= Persona('Marco',35)
```

```python
marco.double(23)
```

```python
marco.quadratoanni()
```
